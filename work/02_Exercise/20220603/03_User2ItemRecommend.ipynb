## 手順

## 注意点

## アイデア例

1. ログインユーザに対して、過去の視聴履歴から評価の高かったアニメを取得し、「もう一度見たいアニメ」としてトップページに表示する【難易度: 低】
1. ログインユーザに対して、過去の視聴履歴から好みのジャンルを取得し、それぞれのジャンルについて、まだ見たことのないおすすめのアニメをトップページに表示する【難易度: 中】
1. ログインユーザに対して、高評価をつけたアニメを取得し、そのアニメの類似アニメをメールでおすすめする【難易度: 中】
1. ログインユーザに対して、過去の視聴履歴から傾向の似ている他のユーザを取得し、他の類似ユーザが視聴していて対象ログインユーザがまだ見ていないアニメをメールでおすすめする【難易度: 高】

## ヒント

- ユーザに対してアニメをおすすめする際、すでに視聴済みのアニメはレコメンドから除かれているほうが望ましいことが多いです
- ユーザからおすすめアイテムを取得する方法として、「まずユーザの既知のアイテム群を取得し、そのアイテムに類似する未知のアイテムを取得する方法」（Item to Itemの応用や**アソシエーション分析**【難易度: 中】）と、「対象ユーザと類似するユーザ群を取得し、そのユーザ群が嗜好しているアイテムの中から未知のアイテムを取得する方法」（**協調フィルタリング**【難易度: 高】）があります
- アソシエーション分析の古典的なアルゴリズムとして、**アプリオリ**アルゴリズムがあり、「アイテムの組み合わせ（A, B, C, ...）を買った人のn%は、アイテムの組み合わせ（X, Y, Z, ...）を買う」といった**相関ルール**を抽出できます
- 相関ルールの評価指標として、**支持度（support）** や **確信度（confidence）**、 **リフト（lift）** などがあります。
- 実用的に利用されているUser to Itemレコメンドの多くは、機械学習による反復計算や再帰計算が必要なアルゴリズムであり、SQLだけで実現するには不向きなため、SQLでは前処理だけを行い、レコメンドの計算はライブラリを利用するケースが多くあります。しかし、今回のワークショップでは、既存のライブラリの使い方を学ぶのではなく、ゼロから自身の手でレコメンドのアルゴリズムを理解することを目標としているため、SQLだけで実現できるロジックを考えてみてください。
