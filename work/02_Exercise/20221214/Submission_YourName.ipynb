  - こちらの取引履歴データは、[国税庁：暗号資産に関する税務上の取扱いについて(FAQ)](https://www.nta.go.jp/publication/pamph/pdf/virtual_currency_faq_03.pdf)の事例をもとに作成しています

引用
```
（例） 3月1日に初めてビットコインを購入して以降、内訳のとおり、数度にわたり購入
と売却を行い、1年間の売却額（数量）の総額は、5,295,000 円（5 BTC）、購入額（数
量）の総額は、4,037,800 円（6.5 BTC）でした。
（内訳）
・ 3月1日 4 BTC を 1,845,000 円で購入（保有数量 4 BTC）

・ 7月10日 2 BTC を 2,400,000 円で売却（保有数量 4 BTC）

・ 11月30日 3 BTC を 2,895,000 円で売却（保有数量 1.5 BTC）
```

````python
%%sql
CREATE SCHEMA IF NOT EXISTS cryptocurrency;

DROP TABLE IF EXISTS cryptocurrency.btc_trade;
CREATE TABLE IF NOT EXISTS cryptocurrency.btc_trade (
  trade_time TIMESTAMP,
  trade_type TEXT,
  trade_amount NUMERIC,
  btc_amount NUMERIC,
  jpy_amount NUMERIC
);

INSERT INTO cryptocurrency.btc_trade VALUES
  ('2018-03-01', '買い', '1845000', '4', '-1845000')
, ('2018-06-20', '買い', '1650000', '2', '-1650000')
, ('2018-07-10', '売り', '2400000', '-2', '2400000')
, ('2018-09-15', '買い', '542800', '0.5', '-542800')
, ('2018-11-30', '売り', '2895000', '-3', '2895000')
;
````

### 問1. 暗号資産の取得原価計算

- 各取引の終了時におけるBTCの残高(btc_balance)を計算してください
- 各取引の終了時における 1 BTC あたりの取得原価、および、BTC残高の評価額について、総平均法(gross_avg)および移動平均法(moving_avg)による方法でそれぞれ計算してください
- 総平均法および移動平均法の詳細については、[国税庁：暗号資産に関する税務上の取扱いについて(FAQ)](https://www.nta.go.jp/publication/pamph/pdf/virtual_currency_faq_03.pdf)のpp.14-16に詳細が記載されています

引用
```
総平均法： 同じ種類の暗号資産について、年初時点で保有する暗号資産の評価額とそ
の年中に取得した暗号資産の取得価額との総額との合計額をこれらの暗号資
産の総量で除して計算した価額を「年末時点での１単位当たりの取得価額」
とする方法をいいます。

移動平均法： 同じ種類の暗号資産について、暗号資産を取得する都度、その取得時点に
おいて保有している暗号資産の簿価の総額をその時点で保有している暗号資
産の数量で除して計算した価額を「取得時点の平均単価」とし、その年12月31
日から最も近い日において算出された「取得時点の平均単価」を「年末時点で
の１単位当たりの取得価額」とする方法をいいます。
```
